```python
corpus = np.array(['que dia es hoy', 'martes el dia de hoy es martes', 'martes muchas gracias'])
```

### 1 - Obtener el vocabulario del corpus (los términos utilizados)
- Cada documento transformarlo en una lista de términos
- Armar un vector de términos no repetidos de todos los documentos

```python
#creo una lista de palabras para cada documento
doc0= list(corpus[0].split())
print (doc0)


doc1= list(corpus[1].split())
print (doc1)

doc2= list(corpus[2].split())
print (doc2)

```

```python
#concateno las 3 listas
doc0.extend(doc1)
doc0.extend(doc2)
print (doc0)
```

```python
#me quedo con los valores unicos luego de la concatenacion
import pandas as pd
vocab_completo=[]
vocab_completo = pd.unique(doc0)
```

### 2- OneHot encoding
Data una lista de textos, devolver una matriz con la representación oneHotEncoding de estos

```python
#creo una lista de palabras para cada documento
docu0= list(corpus[0].split())
print (docu0)

docu1= list(corpus[1].split())
print (docu1)

docu2= list(corpus[2].split())
print (docu2)

lst = []

lst.append(docu0)
lst.append(docu1)
lst.append(docu2)
print(lst)
```

```python
print(lst)
```

```python
oneHotEncoding = np.zeros((len(lst), len(vocab_completo)))

for i, doc in enumerate(lst):
  for j, term in enumerate(vocab_completo):
    oneHotEncoding[i,j] = term in doc

print(oneHotEncoding)
```

### 3- Vectores de frecuencia
Data una lista de textos, devolver una matriz con la representación de frecuencia de estos

```python
tf = np.zeros((len(lst), len(vocab_completo)))

for i, doc in enumerate(lst):
  for j, term in enumerate(vocab_completo):
    tf[i,j] = doc.count(term)

print(tf)
```

```python
vocab_completo = np.sum(lst)
diccionario = {}
for palabra in vocab_completo:
  if palabra in diccionario:
    diccionario[palabra] += 1
  else:
    diccionario[palabra] = 1

for palabra in diccionario:
  frecuencia = diccionario[palabra]
  print(f"La palabra '{palabra}' tiene una frecuencia de {frecuencia}")
```

### 4- TF-IDF
Data una lista de textos, devolver una matriz con la representacion TFIDF

```python
df = oneHotEncoding.sum(axis=0)
n = len(lst)
idf = np.log10(n/df)
tf_idf = (tf * idf).round(decimals=2)
print(tf_idf)
```
